```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import dynamics as Sim
%matplotlib inline

A = np.matrix('-3, -2; -1, -3')
B = np.matrix('6;6')
sim1 = Sim.NC(A,B)
x_NCE,y_NCE=sim1.get_NCE_prices()
x_PCE,y_PCE=sim1.get_PCE_prices()
```

```python
##USER INPUT HERE##
xmin,xmax=.4,1
ymin,ymax=.4,1
manual_locations =""# [(x_NCE, y_NCE), (x_PCE, y_PCE),()]

title="Contours of Produer Welfare ($)"
x_axis_label="Player 1 Price ($)"
y_axis_label="Player 2 Price ($)"
nce_label="NCE"
pce_label="PCE"
levels=np.arange(3.4,4,0.1)
```

```python
#Build and display the contour graph
p1,p2,sw=sim1.gen_contour_graph_data(xmin,xmax,ymin,ymax)
plt.figure()
CS = plt.contour(p1, p2, sw,6,colors='k',levels=levels)
plt.plot([x_NCE],[y_NCE],marker="o",markersize=4)
plt.plot([x_PCE],[y_PCE],marker="o",markersize=4)
plt.clabel(CS, inline=1, fontsize=10)

plt.annotate(
        nce_label,
        xy=(x_NCE, y_NCE), xytext=(-20, 0),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='blue', alpha=0.5),
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))

plt.annotate(
        pce_label,
        xy=(x_PCE, y_PCE), xytext=(-20, -20),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='green', alpha=0.5),
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))

plt.xlabel(x_axis_label)
plt.ylabel(y_axis_label)
plt.title(title)
```
